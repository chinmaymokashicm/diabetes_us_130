```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned.csv")
df.head()
```

```python
df.columns
```

```python
# Looping through X and Y variables
list_X = ["gender", "race", "age"]
list_Y = ["readmitted", "change", "insulin", "A1Cresult"]
for X in list_X:
    for Y in list_Y:
        # pd.crosstab(df[X], df[Y]).apply(lambda r: r/r.sum(), axis=1).plot(kind="barh")
        df_crosstab = pd.crosstab(df[X], df[Y], normalize="index", margins=False)
        # sns.heatmap(df_crosstab, cmap="rocket_r", annot=True, fmt="g")
        sns.heatmap(df_crosstab, cmap="rocket_r", annot=True, fmt=".2%")
        plt.show()
```

```python
df_crosstab = pd.crosstab(df.A1Cresult, df.readmitted, margins=False)
sns.heatmap(df_crosstab, cmap="rocket_r", annot=True, fmt="g")
plt.show()
```
